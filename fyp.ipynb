```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Separate the features (X) and the labels (y)
X = sampled_data.drop('label', axis=1)  # Drop the 'label' column to get features
y = sampled_data['label']  # The 'label' column is the target

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Combine the resampled features and labels into a new DataFrame
resampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['label'])], axis=1)

# Display the shape of the resampled data to verify
print("Original sampled data shape:", sampled_data.shape)
print("Resampled data shape:", resampled_data.shape)

# Count occurrences of each class in the resampled dataset
print(resampled_data['label'].value_counts())

# Optionally, save the resampled dataset to a new CSV file
resampled_data.to_csv('resampled_data.csv', index=False)

# Plot the occurrences of each attack type after applying SMOTE
plt.figure(figsize=(10, 6))
resampled_data['label'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Occurrences of Each Attack Type After SMOTE')
plt.xlabel('Attack Type')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a new column 'Attack' based on the 'label' column
resampled_data['Attack'] = resampled_data['label'].apply(lambda x: 'Non-Attack' if x == 'BenignTraffic' else 'Attack')

# Display the first few rows to verify the changes
print(resampled_data[['label', 'Attack']].head())

# Optionally, save the updated dataset with the new binary classification
resampled_data.to_csv('binary_classified_data.csv', index=False)

# Count occurrences of each category in the new 'Attack' column
print(resampled_data['Attack'].value_counts())

# Plot the occurrences of Attack and Non-Attack
plt.figure(figsize=(6, 4))
resampled_data['Attack'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Number of Non-Attack vs Attack Occurrences')
plt.xlabel('Category')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```
